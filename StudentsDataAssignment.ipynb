```python


# Assuming your DataFrame is named 'df'
Numeric_cols=['age','Medu','Fedu','traveltime','studytime','failures','freetime','goout','health','absences','1st Grade','2nd Grade','Final Grade']

for column in Numeric_cols:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1

    # Identify outliers
    outliers = (df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))

    # Display rows with outliers
    print("the outliers are:",df[outliers])

```

```python
numbers=['age','failures','absences','health','1st Grade','2nd Grade','Final Grade']
for cols in numbers:
    plt.hist(df[cols],bins=20,color="green")
    plt.xlabel(cols)
    plt.ylabel('Frequency')
    plt.title(f'Histogram of {cols}')
    plt.show()
```
